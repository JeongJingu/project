## 상관분석

```python
'''
1. 전체 컬럼 상관분석
2. 지역별 날씨 컬럼과 가장 많은 배달의 상관분석
3. 강수 유형별 날씨 컬럼과 가장 많은 배달의 상관분석
* 기본적으로 년도 나누어서 하기
'''
```

### 1. 서울, 경기, 인천 통합

### 1). 컬럼 별 상관분석

```python
a = ['광역시도명','시군구명','시간대별 시간','강수 유형명','습도값','강수량 값','기온값','풍속값','년도','월','일','요일','계절', '불쾌지수', '가장 많은 배달']
total_1 = data[a]

## 상관분석
x = total_1.corr()
# print(x)

## 각 컬럼별 0.5 이상 혹은 -0.5 이하인 것 추출
for i in a:
    r = x[i]
    r = r.drop(index = i)
    t = r[r>=0.5]
    print(f'{i} 컬럼 상관분석 0.5 이상')
    print(t)
    print('-'*80)

## 결과
'''
1. 강수 유형과 강수량값은 서로 상관 관계 있음
2. 기온값과 불쾌지수는 아주 강한 상관 관계 있음
3. 모델링 하는데 더욱 필요로 하는 강수 유형과 기온값을 채택 
'''
```

### 2). 최종 모델링에 사용 되었던 컬럼 별 상관분석

```python
a = ['시군구명','시간대별 시간','강수 유형명','습도값','기온값','풍속값',
     '년도','월','일','요일','계절','가장 많은 배달']
total_2 = data[a]

## 상관분석 
x = total_2.corr()

## 각 컬럼별 0.5 이상 혹은 -0.5 이하인 것 추출
for i in a:
    r = x[i]
    r = r.drop(index = i)
    t = r[r>=0.5]
    print(f'{i} 컬럼 상관분석 0.5 이상')
    print(t)
    print('-'*80)
    
## 결과 
'''
1. 각 컬럼별 상관 관계가 있는 것이 없음 모든 컬럼 사용 
'''
```

### 3). 지역별 상관분석

```python
## 1. 서울
a = ['시군구명','시간대별 시간','강수 유형명','습도값','기온값','풍속값','년도','월','일','요일','가장 많은 배달']
se = data[data['광역시도명'] == 1][a]

# 상관분석 - 2019
se_2019 = se[se['년도'] == 2019]
region_se_2019 = se_2019['시군구명'].unique()
for i in region_se_2019:
    x = se_2019[se_2019['시군구명'] == i]
    x = x.drop(columns = ['시군구명', '년도'])
    y = x.corr()
    print(f'서울_2019_{i}지역 상관분석')
    print(y)
    print('-'*80)

    
# 상관분석 - 2020
se_2020 = se[se['년도'] == 2020]
region_se_2020 = se_2020['시군구명'].unique()
for i in region_se_2020:
    x = se_2020[se_2020['시군구명'] == i]
    x = x.drop(columns = ['시군구명','년도'])
    y = x.corr()
    print(f'서울_2020_{i}지역 상관분석')
    print(y)
    print('-'*80)
```
